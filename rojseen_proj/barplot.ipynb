```python
import matplotlib.pyplot as plt
import numpy as np

# Data
storeys = ['4', '8', '16']
esm_values = [3575, 12294, 30112]
rsm_values = [3252, 16406, 40388]

# Position of bars on X axis
x = np.arange(len(storeys))  # [0, 1, 2]
bar_width = 0.35

# Plotting
fig, ax = plt.subplots()

# ESM bars
esm_bars = ax.bar(x - bar_width/2, esm_values, width=bar_width, label='ESM', color='skyblue')
# RSM bars
rsm_bars = ax.bar(x + bar_width/2, rsm_values, width=bar_width, label='RSM', color='salmon')

# Labels and title
ax.set_xlabel('Storey Levels')
ax.set_ylabel('Base Shear (in kN)')
ax.set_title('Base Shear Variation with Storey Levels : ESM vs. RSM (X-direction)')
ax.set_xticks(x)
ax.set_xticklabels(storeys)
ax.legend()

# Optional: Add values on top of bars
for bars in [esm_bars, rsm_bars]:
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height}',
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  # Offset
                    textcoords="offset points",
                    ha='center', va='bottom')
plt.ylim(0, 45000)  # Increase max y-limit above highest label (256.0)
# plt.text(bar_x, bar_height + 5, f"{bar_height}", ha='center')
plt.tight_layout()
plt.show()
```
